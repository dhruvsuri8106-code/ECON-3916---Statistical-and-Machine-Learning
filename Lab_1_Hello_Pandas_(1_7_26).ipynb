```python
#Project 1 - Dhruv Suri

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = {
  'Country': ['United States', 'Argentina', 'Australia', 'Brazil', 'Britain', 'China',
        'Egypt', 'Euro area', 'Hong Kong', 'Indonesia', 'Japan', 'Mexico',
        'Norway', 'Pakistan', 'Philippines', 'Russia', 'Saudi Arabia',
        'South Africa', 'South Korea'],
  'iso_a3': [
        'USA', 'ARG', 'AUS', 'BRA', 'GBR', 'CHN',
        'EGY', 'EUZ', 'HKG', 'IDN', 'JPN', 'MEX',
        'NOR', 'PAK', 'PHL', 'RUS', 'SAU',
        'ZAF', 'KOR'
    ],
    'local_price': [
        4.79, 28.00, 5.30, 13.50, 2.89, 17.20,
        16.93, 3.68, 18.80, 27939, 370, 49.00,
        48.00, 300, 163, 89.00, 11.00,
        25.50, 4100
    ],
    'dollar_ex': [
        1.00, 8.62, 1.23, 2.59, 0.66, 6.21,
        7.36, 0.86, 7.74, 12472, 117.83, 14.63,
        7.62, 100.67, 44.41, 65.44, 3.75,
        11.49, 1084.66
    ]





}
```

```python
df= pd.DataFrame(data)
```

```python
df
```

```python
#calculate price in USD by dividing the local price by the dollar multiplier

df['dollar_price'] = df['local_price'] / df['dollar_ex']

us_price = df.loc[df['Country'] == 'United States', 'dollar_price'].values

df['implied_price'] = df['dollar_price'] / us_price

df['valuation'] = df['implied_price'] - df['dollar_ex']/df['dollar_ex']

df ['valuation_pct'] = df['valuation'] * 100
```

```python
df
```

```python
plt.figure

#bar plot
sns.barplot(x= 'valuation_pct', y= 'Country', data= df.sort_values('valuation_pct'))
```
